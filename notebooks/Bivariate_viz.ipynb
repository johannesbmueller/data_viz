```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Load the iris dataset
iris = sns.load_dataset("iris")
```

```python
# Preview the data
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris)
```

```python
sns.regplot(x='sepal_length', y='petal_length', data=iris)
```

```python
plt.hexbin(x=iris['sepal_length'], y=iris['petal_length'], gridsize=25, cmap='Blues')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar(label='Counts')
```

```python
sns.kdeplot(x='sepal_length', y='petal_length', data=iris, fill=True)
```

```python
sns.jointplot(x='sepal_length', y='petal_length', data=iris, kind='scatter')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length')
plt.title('Scatter Plot: Sepal vs Petal Length')
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Scatter Plot with Color (Hue) by Species')
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', style='species')
plt.title('Scatter Plot with Shape (Style) by Species')
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', style='species')
plt.title('Scatter Plot with Color + Shape by Species')
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', size='sepal_width', hue='species')
plt.title('Scatter Plot with Size = Sepal Width')
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', alpha=0.6)
plt.title('Scatter Plot with Transparency')
plt.show()
```

```python
g = sns.FacetGrid(iris, col='species')
g.map_dataframe(sns.scatterplot, x='sepal_length', y='petal_length')
g.fig.suptitle('Faceted Scatter Plots by Species', y=1.05)
plt.show()
```
