```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Load the iris dataset
iris = sns.load_dataset("iris")
```

```python
# Preview the data
iris.head()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(iris['sepal_length'], bins=15, kde=False)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(iris['sepal_length'], fill=True)
plt.title('KDE Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(2, 6))
sns.boxplot(y=iris['sepal_length'])
plt.title('Box Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(2, 6))
sns.violinplot(y=iris['sepal_length'])
plt.title('Violin Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(4, 6))
sns.swarmplot(x=np.zeros(len(iris)), y=iris['sepal_length'], size=4)
plt.title('Swarm Plot of Sepal Length')
plt.ylabel('Sepal Length (cm)')
plt.xticks([])  # Remove x-axis labels
plt.show()
```

```python
plt.figure(figsize=(6, 4))
stats.probplot(iris['sepal_length'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Sepal Length')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.show()
```
